# if ~ elif ~ else (다중선택 조건문)

- 조건이 여러 개일 경우 if 다음에 elif 를 쓴다.
- elif는 여러개 쓸 수 있고 앞의 조건이 거짓일 때만 실행
- else는 제일 마지막 위의 조건들이 다 거짓일 때 사용

```python
score = int(input('점수를 입력해주세요:'))

if score >= 90:
  print('A')
elif score >= 80:
  print('B')
elif score >= 70:
  print('C')
else :
  print('D')
```

사용자로 부터 점수를 입력받아, 합격 또는 과락으로 불합격 또는 불합격을 출력하는 프로그램

- 영어, 수학, 국어 3개 과목에서,
- 평균 60점 이상이면 합격,
- 단, 한과목이라도 40점 미만이라면 과락(불합격)으로 처리

```python
english = int(input('영어 점수를 입력하세요 : '))
math = int(input('수학 점수를 입력하세요 : '))
language = int(input('국어 점수를 입력하세요 : '))

score = (english + math + language) / 3

if score >= 60 :
  print('합격')
elif english < 40 :
  print('과락(불합격)')
elif math < 40 :
  print('과락(불합격)')
elif language < 40 :
  print('과락(불합격)')
else :
  print('불합격')

# 내꺼
```

```python
eng = int(input('당신의 영어 점수는 : '))
math = int(input('당신의 수학 점수는 : '))
kor = int(input('당신의 국어 점수는 : '))

avg = (eng + math + kor) / 3

# 불합격
if avg < 60 :
  print('평균 미달로 불합격입니다.')

# 과락으로 불합격인 경우 :
elif eng < 40 or math < 40 or kor < 40:
  print('아쉽지만, 평균은 60점이상인데 과락이 있어 불합격이네요.')

# 합격인 경우
else :
  print('합격입니다!')
```

# while 조건 반복문
- 조건이 참이면 계속 반복되는 구문
- while 두에 조건을 쓰고 콜론을 붙인다.
- 붙여쓰기 후에 조건이 참일 경우 실행할 명령문을 적는다.
- 조건인 거짓일 경우, 중단할 조건을 추가한다.(그렇지 않으면 무한루프에 빠지게 된다.)

```python
'''
1) 변수 선언
2) while 조건식
3) 조건의 변경식
'''

import time

x = 10

while x > 0 :
  print(x, '는(은) 0보다 큰 수 입니다.')
  x -= 1
  time.sleep(1.5) # 코드를 괄호안의 입력값만큼 잠시 멈춰주는 함수 (초 단위)
```

```python
# while 문으로 1부터 10까지 합을 구하는 코드 작성

i = 1
sum = 0 # 총 더해진 값이 저장될 변수라서 0으로 둔다.

while i <= 10 :
  sum = sum + i
  i += 1
  print('총합 : ', sum)
```

```python
# 인덱스 지정

df = pd.DataFrame({'이름': ['손흥민', '메시', '호날두'],
                   '나이': [30, 37, 35],
                   '소속팀': ['토트넘', '마이애미', '알나스']},
                  index = [1, 2, 3])
df
```

```python
# 리스트 안에 리스트 -> 컬럼 이름을 정하기(변수명 직접 지정)
df = pd.DataFrame([['손흥민', 30, '토트넘'],
                   ['메시', 37, '마이애미'],
                   ['호날두', 35, '알나스']],
                  columns = ['이름', '나이', '소속'],
                  index = ['한국', '아르헨티나', '포르투갈'])
df
```

```python
# 기존의 df에서 컬럼 이름을 변경해주고 싶은 경우
df.columns = ['name', 'age', 'club']
df
```

```python
# 특정 인덱스랑 변수명(컬럼)만 바꿔주고 싶은 경우

# 1. club을 team 변경
df = df.rename(columns = {'club': 'team'}) # rename 함수가 동작 했을 때만 변수명이 뱌뀐다. 그래서 df로 변수 지정을 해준다.
df
```

```python
# 2. 인덱스 한국을 0으로 변경
df = df.rename(index = {'한국': 0})
df
```

위의 rename 함수 실행 결과를 저장하려면 변수명으로 저장을 해주면 된다.
